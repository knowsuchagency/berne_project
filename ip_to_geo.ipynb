```python
from collections import namedtuple
from pprint import pprint
import csv
import gmplot
import geoip2.database
```

```python
GeoIP = namedtuple("GeoIP", ['ip', 'city', 'state', 'lat', 'lon'])
```

```python
ip_addresses = []
with open("ip_addresses.csv") as infile:
    csv_reader = csv.reader(infile)
    for row in csv_reader:
        ip_addresses.append(row[0])
```

```python
db_reader = geoip2.database.Reader('geolite.mmdb')
```

```python
Geos = []
for ip in ip_addresses:
    r = db_reader.city(ip)
    Geos.append(GeoIP(
        ip=ip,
        city=r.city.name,
        state=r.subdivisions.most_specific.name,
        lat=r.location.latitude,
        lon=r.location.longitude
        ))
```

```python
with open('ip_geos.csv', 'w') as outfile:
    writer = csv.DictWriter(outfile, fieldnames=GeoIP._fields)
    writer.writeheader()
    writer.writerows([g._asdict() for g in Geos])
```

```python
for geo in Geos:
    print(geo)
```
